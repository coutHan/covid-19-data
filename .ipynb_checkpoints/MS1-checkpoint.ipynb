## Step 1: import the dataset to a pandas dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as dates
import numpy as np
import pylab 
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.graphics.gofplots import qqplot_2samples
import math
import copy
import warnings
import os
import csv
warnings.filterwarnings('ignore')

##Load data
dataUS = pd.read_csv("us.csv")
dataSize = len(dataUS.index)
dataStates = pd.read_csv("us-states.csv")
dataStateSize = len(dataStates.index)
# print(dataUS.loc[0], dataSize, "\n", dataStates.loc[0] ,dataStateSize)
# if (dataUS.loc[1]["date"]>dataUS.loc[0]["date"]):
#     print("nice")
```

## Step 2. Data cleaning and reorganization
###  Step 2.1: Extract US data (daily new death from total death) and save to a file 

```python
## US data
usDeath = []
for i in range(dataSize):
    if (i - 1) < 0:
        usDeath.append([dataUS.loc[i]["date"],dataUS.loc[i]["deaths"]])
    else:   
        usDeath.append([dataUS.loc[i]["date"],dataUS.loc[i]["deaths"]-dataUS.loc[i-1]["deaths"]])
with open("clean-data/US.csv", 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(usDeath)

#print (usDeath)
```

### Step 2.2: Extract states' COVID data and save to files

```python
# State data
statesDate = {}
statesData = {}
for i in range(dataStateSize):
    if (dataStates.loc[i]["state"] in statesDate):
        statesData[dataStates.loc[i]["state"]].append([dataStates.loc[i]["date"],dataStates.loc[i]["deaths"]-statesData[dataStates.loc[i]["state"]][-1][1]])
    else:
        statesDate[dataStates.loc[i]["state"]] = dataStates.loc[i]["date"]
        statesData[dataStates.loc[i]["state"]] = []
        # First COVID case
        statesData[dataStates.loc[i]["state"]].append([dataStates.loc[i]["date"],dataStates.loc[i]["deaths"]])
for key in statesData:
    path = "clean-data/" + key + ".csv"
    print(path)
    with open(path, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(statesData[key])
    
## Get latest date of COVID 
#startDate = statesDate[max(statesDate,key=statesDate.get)]
#print(statesData)
```

## Q5. (1.4%) Star plot.
### We want to compare the number of Facebook likes for the following eight countries: Australia, Canada, France, Germany, India, Spain, UK, and USA. 
### Specifically, we want to draw three curves, each representing one of the following values for all countries: mean of movie_facebook_likes,mean of actor_1_facebook_likes, mean of director_facebook_likes
### Generate a star plot that shows the above information. Each of the equiangular spokes should represent one country.

```python
##TODO
# 1st dimension: Australia, Canada, France, Germany, India, Spain, UK, and USA, 
# 2nd dimension: movie_facebook_likes, actor_1_facebook_likes,director_facebook_likes, number of movies (exclude movies with nan value)
dataFacebook = [[copy.deepcopy(0) for x in range(4)] for y in range(8)]
countryMap = {
    "Australia": 0,
    "Canada": 1,
    "France": 2,
    "Germany": 3,
    "India": 4,
    "Spain": 5,
    "UK": 6,
    "USA": 7,
}
for i in range(dataSize):
    if (data.loc[i]["country"] in countryMap):
        if (math.isnan(data.loc[i]["movie_facebook_likes"])) or (math.isnan(data.loc[i]["actor_1_facebook_likes"])) or (math.isnan(data.loc[i]["director_facebook_likes"])):
            pass
        else:
            countryLoc = countryMap[data.loc[i]["country"]]
            dataFacebook[countryLoc][0] += data.loc[i]["movie_facebook_likes"]
            dataFacebook[countryLoc][1] += data.loc[i]["actor_1_facebook_likes"]
            dataFacebook[countryLoc][2] += data.loc[i]["director_facebook_likes"]
            dataFacebook[countryLoc][3] += 1
    else:
        pass 
# 1st dimension: Australia, Canada, France, Germany, India, Spain, UK, and USA, 
# 2nd dimension: mean of movie_facebook_likes, mean of actor_1_facebook_likes, mean of director_facebook_likes
dataMeanFacebook = [[copy.deepcopy(0) for x in range(3)] for y in range(8)]
for i in range(8):
    dataMeanFacebook[i][0] = dataFacebook[i][0]/dataFacebook[i][3]
    dataMeanFacebook[i][1] = dataFacebook[i][1]/dataFacebook[i][3]
    dataMeanFacebook[i][2] = dataFacebook[i][2]/dataFacebook[i][3]
# 1st dimension: mean of movie_facebook_likes, mean of actor_1_facebook_likes, mean of director_facebook_likes
# 2nd dimension: Australia, Canada, France, Germany, India, Spain, UK, and USA,
dataMeanFacebookT = np.array(dataMeanFacebook).T.tolist()
dataMeanFacebookT[0].append(dataMeanFacebookT[0][0])
dataMeanFacebookT[1].append(dataMeanFacebookT[1][0])
dataMeanFacebookT[2].append(dataMeanFacebookT[2][0])
#print(dataMeanFacebookT)
# #thetas = np.linspace(0, 2*np.pi, num=8,endpoint=False)
thetas = [0.0, 0.78539816, 1.57079633, 2.35619449, 3.14159265, 3.92699082, 4.71238898, 5.49778714, 0.0]
ax = plt.subplot(111, projection='polar')
# Plot series data and Legend
ax.plot(thetas, dataMeanFacebookT[0], c='blue', label="movies likes")
ax.plot(thetas, dataMeanFacebookT[1], c='orange', label="1st actor likes")
ax.plot(thetas, dataMeanFacebookT[2], c='green', label="director likes")
ax.set_xticklabels(["Australia", "Canada", "France", "Germany", "India", "Spain", "UK", "USA"])
ax.set_theta_zero_location('N')
plt.legend(bbox_to_anchor=(1, 0, 0.5, 1),loc="upper right")
plt.show()
```

```python
##TODO 
insights = 'Star plot displays multivariate data in the form of a two-dimensional chart of multiple quantitative variables represented on axes starting from the same point. In our star plot of the number of Facebook likes for the eight countries, we can see that Australia has the highest number of average movie Facebook likes, the USA has the highest number of average 1st actor Facebook likes, and Spain has the highest number of director Facebook likes.'
print('Insights:\t', insights)
```
